```python
import numpy as np

def f(x1,x2):
    return (x1**2 / 2) + x2**2

def grad_f(x1,x2):
    return (x1, 2*x2)

# Initial point.
x0 = np.array((1,1))

# Global minimum.
x_star = np.array((0,0))

# Define the error threshold.
err_thresh = 0.01

# Lipschitz constant of the gradient of f.
L = 2

# Since f is L smooth, lr = 1/L is the optimal learning rate.
lr = 1/(L)  

# Initialise the iteration count.
iter_count = 0

# Initialise the current point.
xi = x0.copy()

# Initialise the starting error.
err = np.linalg.norm(xi - x_star)**2
while err > err_thresh:
    # Compute the gradient at the current point.
    grad = grad_f(*xi)

    # Update the current point.
    x1 = xi[0] - lr*grad[0]
    x2 = xi[1] - lr*grad[1]
    xi = (x1, x2)

    # Compute the error.
    err = np.linalg.norm(xi - x_star)
    
    iter_count += 1

    print(f'iter: {iter_count}, xi: {xi}, f(xi): {f(*xi)}, err: {err}')
```
